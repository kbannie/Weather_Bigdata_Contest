### 변수 설명
1. fc_year, month, day, hour : 기준 발표 시각
2. ef_year, month, day, hour : 예측 시각
3. dh : 기준 시각 - 예측 시각
    * dh는 3h, 6h, 9h, ... , 240h. 3시간 간격으로 총 80개
    * 기준 발표 시각이 5월 1일 오전 9시인 경우, 예측 시각은  5월 1일 오전 9시에서 3시간후, 6시간 후, ... , 240시간 후로 총 80개의 강수 누적 확률 예측
4. stn4contest : AWS 지점 코드
5. v01 ~ v09 : 강우 누적 확률(앙상블 모델의 구간별 강수량 누적 확률 값)
6. vv : 실제 강수량(지점별 1시간 강수 자료의 3시간 누적 값)
7. class_interval : 강수계급

### 전처리

```python
# 월 원핫인코딩 변수 생성
df_fc_month_oh = pd.get_dummies(df['fc_month'], prefix='fc', dtype=int)
df_ef_month_oh = pd.get_dummies(df['ef_month'], prefix='ef', dtype=int)
df_dh_oh = pd.get_dummies(df['dh'], prefix='dh', dtype=int)

df_oh = pd.concat([df, df_fc_month_oh, df_ef_month_oh, df_dh_oh], axis=1)
df_oh
```

```python
# 자료 선택
# v01(0.1mm 이상 누적 확률)이 일정값보다 작으면 무조건 무강수일것이라 가정
df_oh_choice = df_oh[df_oh['v01'] > 20]
print(len(df_oh_choice))
```

```python
df_total = df_oh_choice
print(df_total.columns)
```

### 모델: 회귀
* 종속변수: vv
* 목표: 앙상블 확률 자료(v01~v09) 활용하여 3시간 누적 강수량의 계급 구간 예측

```python
# 변수선언
X = df_total[['prob_mean',
              'v01', 'v02', 'v03', 'v04', 'v05', 'v06', 'v07', 'v08', 'v09',
              'v01_w', 'v02_w', 'v03_w', 'v04_w', 'v05_w', 'v06_w', 'v07_w', 'v08_w', 'v09_w',
              'fc_5', 'fc_6', 'fc_7', 'fc_8', 'fc_9', 'ef_5', 'ef_6', 'ef_7', 'ef_8', 'ef_9', 'ef_10',
              'dh_3', 'dh_6', 'dh_9', 'dh_12', 'dh_15', 'dh_18', 'dh_21', 'dh_24',
              'dh_27', 'dh_30', 'dh_33', 'dh_36', 'dh_39', 'dh_42', 'dh_45', 'dh_48',
              'dh_51', 'dh_54', 'dh_57', 'dh_60', 'dh_63', 'dh_66', 'dh_69', 'dh_72',
              'dh_75', 'dh_78', 'dh_81', 'dh_84', 'dh_87', 'dh_90', 'dh_93', 'dh_96',
              'dh_99', 'dh_102', 'dh_105', 'dh_108', 'dh_111', 'dh_114', 'dh_117',
              'dh_120', 'dh_123', 'dh_126', 'dh_129', 'dh_132', 'dh_135', 'dh_138',
              'dh_141', 'dh_144', 'dh_147', 'dh_150', 'dh_153', 'dh_156', 'dh_159',
              'dh_162', 'dh_165', 'dh_168', 'dh_171', 'dh_174', 'dh_177', 'dh_180',
              'dh_183', 'dh_186', 'dh_189', 'dh_192', 'dh_195', 'dh_198', 'dh_201',
              'dh_204', 'dh_207', 'dh_210', 'dh_213', 'dh_216', 'dh_219', 'dh_222',
              'dh_225', 'dh_228', 'dh_231', 'dh_234', 'dh_237', 'dh_240'
              ]]
Y = df_total['vv']

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print(len(X_train))
print(len(X_test))
```

```python
# 스케일링
from sklearn.preprocessing import StandardScaler, MinMaxScaler, PowerTransformer

scaling_columns = ['prob_mean', 'v01_w', 'v02_w', 'v03_w', 'v04_w', 'v05_w', 'v06_w', 'v07_w', 'v08_w', 'v09_w']
scaler = StandardScaler()

train_scaled = pd.DataFrame(scaler.fit_transform(X_train[scaling_columns]), columns=scaling_columns, index=X_train.index)
test_scaled = pd.DataFrame(scaler.transform(X_test[scaling_columns]), columns=scaling_columns, index=X_test.index)

train_remaining = X_train.drop(columns=scaling_columns)
test_remaining = X_test.drop(columns=scaling_columns)

X_train_scaled = pd.concat([train_scaled, train_remaining], axis=1)
X_test_scaled = pd.concat([test_scaled, test_remaining], axis=1)
```

```python
from lightgbm import LGBMRegressor
LightGBM = LGBMRegressor(
    num_iterations=200,
    learning_rate=0.1,
    max_depth=5,
    min_data_in_leaf=20,
    num_leaves=31,
    boosting='gbdt',
    bagging_fraction=1.0,
    feature_fraction=1.0,
    lambda_l1=0.0,
    lambda_l2=0.0,
    random_state=42
    )

from sklearn.model_selection import cross_validate
cv = cross_validate(LightGBM, X_train_scaled, Y_train, cv=5, scoring='r2')
```

```python
cv
```

```python
LightGBM.fit(X_train_scaled, Y_train)
```

```python
Y_pred_train = LightGBM.predict(X_train_scaled)

from sklearn.metrics import mean_squared_error, r2_score
RMSE_train = np.sqrt(mean_squared_error(Y_train, Y_pred_train))
R2_train = r2_score(Y_train, Y_pred_train)

print(f'RMSE train: {RMSE_train}')
print(f'R2 train: {R2_train}')
```

```python
Y_pred = LightGBM.predict(X_test_scaled)

from sklearn.metrics import mean_squared_error, r2_score
RMSE = np.sqrt(mean_squared_error(Y_test, Y_pred))
R2 = r2_score(Y_test, Y_pred)

print(f'RMSE: {RMSE}')
print(f'R2: {R2}')
```

```python
# 스케일링
from sklearn.preprocessing import StandardScaler, MinMaxScaler, PowerTransformer

scaling_columns = ['prob_mean', 'v01_w', 'v02_w', 'v03_w', 'v04_w', 'v05_w', 'v06_w', 'v07_w', 'v08_w', 'v09_w']
scaler = StandardScaler()

train_scaled = pd.DataFrame(scaler.fit_transform(X_train[scaling_columns]), columns=scaling_columns, index=X_train.index)
test_scaled = pd.DataFrame(scaler.transform(X_test[scaling_columns]), columns=scaling_columns, index=X_test.index)

train_remaining = X_train.drop(columns=scaling_columns)
test_remaining = X_test.drop(columns=scaling_columns)

X_train_scaled = pd.concat([train_scaled, train_remaining], axis=1)
X_test_scaled = pd.concat([test_scaled, test_remaining], axis=1)
```
