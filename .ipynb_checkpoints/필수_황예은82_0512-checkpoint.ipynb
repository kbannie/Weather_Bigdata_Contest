### VarianceThreshold

```python
#VarianceThreshold

#임계값 0.5로 먼저 지정
threshold =0.5
selector = VarianceThreshold(threshold)
X_reduced = selector.fit_transform(X)

# 선택된 특성을 확인하기 
selected_features=X.columns[selector.get_support(indices=True)]
print('selected_features 선택된 피처 ',selected_features)

#시각화로 간단하게 살펴보기 

plt.figure(figsize=(12,20))
plt.barh(width= selector.variances_, y=X.columns)
plt.xlabel('Variance')
plt.ylabel('Features')
plt.axvline(x=threshold, color='r', label = 'threshold')
```

```python
#VarianceThreshold

#임계값 0.1로 지정
threshold =0.1
selector = VarianceThreshold(threshold)
X_reduced = selector.fit_transform(X)

# 선택된 특성을 확인하기 
selected_features=X.columns[selector.get_support(indices=True)]
print('selected_features 선택된 피처 ',selected_features)

#시각화로 간단하게 살펴보기 

plt.figure(figsize=(12,20))
plt.barh(width= selector.variances_, y=X.columns)
plt.xlabel('Variance')
plt.ylabel('Features')
plt.axvline(x=threshold, color='r', label = 'threshold')

## 임계값 변경 후 차이가 없음
```

### SFS

```python
#chai Feature Selection을 이용해 X다시 지정
X= df[['X11', 'X12', 'X14', 'X21', 'X29', 'X30', 'X33', 'X34', 'X36', 'X46',
       'X47', 'X48', 'X49', 'X51', 'X53', 'X58', 'X59', 'X63', 'X67', 'X71',
       'X72', 'X74', 'X76', 'X87']]

# 훈련 데이터셋으로 분할
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=111)

# 로지스틱회귀모델
model=LogisticRegression(max_iter=100)

## SFS(모델, 피처선정 어떻게 할건지? best 기타 등등, forward , floating , scoring, cv)
# 전진
sfs_forward=SFS(model, k_features= 'best', forward= True, floating=False, scoring='accuracy', cv=3)
# 후진
sfs_backward=SFS(model, k_features= 'best', forward= False, floating=False, scoring='accuracy', cv=3)
# stepwise
sfs_stepwise=SFS(model, k_features= 'best', forward= True, floating=True, scoring='accuracy', cv=3)

# 각 방법을 3개 모두 다 학습 간단하게 fit
sfs_forward =sfs_forward.fit(X_train, y_train)
sfs_backward =sfs_backward.fit(X_train, y_train)
sfs_stepwise =sfs_stepwise.fit(X_train, y_train)
```

```python
#선택된 특성 출력
print('sfs_forward 선택된 특성')
print(sfs_forward.k_feature_names_)


print('sfs_backward 선택된 특성')
print(sfs_backward.k_feature_names_)


print('sfs_stepwise 선택된 특성')
print(sfs_stepwise.k_feature_names_)

###결과
#선택된 최적의 피쳐 구성은 3가지 방식 모두 동일하게 'pclass', 'age', 'sibsp', 'sex_male', 'embarked_S'로 이루어짐을 확인할 수 있었다.
```

```python
#시각화 진행
fig, ax = plt.subplots(1,3, figsize=(15,7))
ax[0].plot(range(1, len(sfs_forward.subsets_)+1),[sfs_forward.subsets_[i]['avg_score'] for i in sfs_forward.subsets_], marker='o')
ax[0].set_title('sfs_forward')

ax[1].plot(range(1, len(sfs_backward.subsets_)+1),[sfs_backward.subsets_[i]['avg_score'] for i in sfs_backward.subsets_], marker='o')
ax[1].set_title('sfs_backward')

ax[2].plot(range(1, len(sfs_stepwise.subsets_)+1),[sfs_stepwise.subsets_[i]['avg_score'] for i in sfs_stepwise.subsets_], marker='o')
ax[2].set_title('sfs_stepwise')

plt.tight_layout()
plt.show()
```

#### 결과
- 'X11', 'X12', 'X14', 'X21', 'X30', 'X33', 'X34', 'X36', 'X46', 'X51', 'X58', 'X59', 'X63', 'X67', 'X71', 'X72', 'X87'

후진 선택법과 stepwise 선택법을 통해 위와 같은 피쳐 선택

## Model Selection
- 모델 선택하기

```python
X = df[['X11', 'X12', 'X14', 'X21', 'X29', 'X30', 'X33', 'X34', 'X36',
        'X46', 'X47', 'X48', 'X49', 'X51', 'X53', 'X58', 'X59', 'X63',
        'X67', 'X71', 'X72', 'X74', 'X76', 'X87']]
y = df['Bankrupt?']

# 훈련 데이터셋으로 분할
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=111)
```

#### RandomForestClassifier

```python
# 랜덤 포레스트 분류 모델 정의
model = RandomForestClassifier()

# 훈련 데이터셋으로 모델 훈련
model.fit(X_train, y_train)

# 테스트 데이터셋으로 예측
y_pred = model.predict(X_test)

# 평가 지표 계산
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# 결과 출력
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

#### SVC

```python
model = SVC()

# 훈련 데이터셋으로 모델 훈련
model.fit(X_train, y_train)

# 테스트 데이터셋으로 예측
y_pred = model.predict(X_test)

# 평가 지표 계산
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# 결과 출력
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

#### LogisticRegression

```python
# 로지스틱 회귀 모델 정의
model = LogisticRegression(max_iter=100)

# 훈련 데이터셋으로 모델 훈련
model.fit(X_train, y_train)

# 테스트 데이터셋으로 예측
y_pred = model.predict(X_test)

# 평가 지표 계산
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# 결과 출력
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

### 결론

정확도: RandomForestClassifier >  LogisticRegression > SVC

따라서 본 데이터 프레임과 피쳐에는 정확도에 있어 큰 차이는 없지만 비교를 해보자면 RandomForestClassifier가 가장 적합하다는 결론을 내릴 수 있었다.
